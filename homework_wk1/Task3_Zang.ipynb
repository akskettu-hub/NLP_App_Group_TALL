```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
url = "https://www.chess.com/news"
response = requests.get(url)
soup = BeautifulSoup(response.content, "html.parser")
```

```python
soup = soup.find("article").find_parent() 

titles = []
for title in soup.find_all("a", class_="post-preview-title"):
    titles.append(title.text.strip())

authors = []
for author in soup.find_all("a", class_="post-preview-meta-username"):
    authors.append(author.text.strip())

dates = []
for date in soup.find_all("div", class_="post-preview-meta-time"):
    date_raw = date.text.replace("\n", "").replace("  ", " ").replace("|", "")
    dates.append(date_raw.strip())
```

```python
news_articles = []
for i in range(len(titles)):
    article = {
        "Title": titles[i],
        "Author": authors[i],
        "Date": dates[i] if i < len(dates) else "No date available"
    }
    news_articles.append(article)


news_articles
```

```python
news_df = pd.DataFrame(news_articles)
news_df
```
